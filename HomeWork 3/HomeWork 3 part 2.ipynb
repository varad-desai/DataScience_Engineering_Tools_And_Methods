```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

```python
y = boston_dataset.target
boston = pd.DataFrame(boston_dataset.data)
```

```python
boston.head()
```

```python
boston.corr()
```

```python
boston1 = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston1.head()
```

```python
boston1.corr()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
correlation_matrix = boston1.corr().round(2)
# In mask, data will not be shown in cells for the upper indices of the squared matrix where mask = true.
mask = np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)] = True
# annot = True to print the values inside the square
with sns.axes_style("white"):
    sns.heatmap(data=correlation_matrix, mask=mask, annot=True)
```

```python
df_y = pd.DataFrame(y)
```

```python
df_y.head()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
#Initialize the linear regression model
reg = linear_model.LinearRegression(fit_intercept=True)
#Split the data into 67% training and 33% testing data
#NOTE: We have to split the dependent variables (x) and the target or independent variable (y)
x_train, x_test, y_train, y_test = train_test_split(boston1, df_y, test_size=0.2, random_state=42)
#Train our model with the training data
reg.fit(x_train, y_train)
#print our price predictions on our test data
y_pred = reg.predict(x_test)
# print(y_pred)
print(y_pred.dtype)
```

```python
print(reg.coef_)
```

The sign of a regression coefficient tells you whether there is a positive or negative correlation between each independent variable the dependent variable. A positive coefficient indicates that as the value of the independent variable increases, the mean of the dependent variable also tends to increase and Vice Versa.


```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# r-squared score of the model
r2 = r2_score(y_test, y_pred)

r2
```

R-squared is a goodness-of-fit measure for linear regression models. This statistic indicates the percentage of the variance in the dependent variable that the independent variables explain collectively. R-squared is a statistical measure of how close the data are to the fitted regression line.

```python
# mean squared error score of the model
mse = mean_squared_error(y_test, y_pred)
mse
```

MSE basically measures average squared error of our predictions. For each point, it calculates square difference between the predictions and the target and then average those values.

```python
# root mean squared error score of the model
mae = mean_absolute_error(y_test, y_pred)
mae
```

In MAE the error is calculated as an average of absolute differences between the target values and the predictions.
